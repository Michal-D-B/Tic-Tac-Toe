```python
from IPython.display import clear_output

def main():
    
    restart = True
    
    while restart:
        
        clear_output()
        p_order = True
        p_ready = False
        dic_board = {
                'row1':[' ',' ',' '],
                'row2':[' ',' ',' '],
                'row3':[' ',' ',' ']
            }
        
        while not p_ready:
            print('Welcome to Tic Tac Toe!')
        
            p1_letter, p2_letter, = player_choice()          
            p_ready = player_ready()
            p_order = player_order(p1_letter)
            clear_output()
            
        display_game(dic_board)
       
        choice_list = ['1','2','3','4','5','6','7','8','9']
        counter = 0
    
        while True:
            counter += 1
            if draw(counter):
                break  
                
            if p_order:
                
                (p1_row, p1_index), p_num_choice = player_number_choice('Player 1:',choice_list)               
                dic_board[p1_row][p1_index] = p1_letter

            elif not p_order:
                
                (p2_row, p2_index),p_num_choice = player_number_choice('Player 2:',choice_list)                      
                dic_board[p2_row][p2_index] = p2_letter
            
            choice_list.remove(p_num_choice)                
            clear_output()                
            display_game(dic_board)
                
            if game_score(dic_board):
                 break
                
            p_order = not p_order
      
        restart = replay()

        
    else:
        clear_output()
        print('*** Thank you for playing! ***')
            


def display_game(dic_board):

    print(f" {dic_board['row1'][0]} | {dic_board['row1'][1]} | {dic_board['row1'][2]} ".upper())
    print('---+---+---')
    print(f" {dic_board['row2'][0]} | {dic_board['row2'][1]} | {dic_board['row2'][2]} ".upper())
    print('---+---+---')
    print(f" {dic_board['row3'][0]} | {dic_board['row3'][1]} | {dic_board['row3'][2]} ".upper())
    
def draw(counter):
    
    if counter == 10:
        print('Draw!')
        return True
    return False

    
def player_choice():

    player = ''
    player2 = ''
    
    while player not in ['x','o']:
        player = input('Player 1 : Do you want to be X or O? ').lower().strip()
        if player not in ['x','o']:
            print('Incorect input, please type X or O! ')
           
    if player == 'x':
        player2 = 'o'
    else:
        player2 = 'x'
        
    return (player, player2)
    

def player_order(player):

    if player == 'x':
        print('Player 1 start first! ')
        return True
    else:
        print('Player 2 start first! ')
        return False

    
def player_ready():

    player = ''

    while player not in ['yes', 'no']:
        player = input('Do you ready to play? Enter Yes or No ').lower().strip()
        
        if player not in ['yes', 'no']:
            print('Wrong input, please enter Yes or No ')
    
    if player == 'yes':
        return True
    else:
        return False

    
def player_number_choice(player,number_list):
    
    choice = '0'
    dic_choice = {
        '1':('row3',0),'2':('row3',1),'3':('row3',2),
        '4':('row2',0),'5':('row2',1),'6':('row2',2),
        '7':('row1',0),'8':('row1',1),'9':('row1',2)
    }
    
    while choice not in number_list:
        choice = input(f'{player} Enter number from 1-9 ').strip()
        if choice not in number_list:
            print('Not in range, please enter number from 1-9')
              
    return dic_choice[choice],choice


def game_score(dic_board):
   
    for z in ['x','o']:       
        for n in range(0,3):
            
            if dic_board['row1'][n] == dic_board['row2'][n] == dic_board['row3'][n] == z:             
                print('Congratulations! You have won the game!')
                return True
                    
            if dic_board[f'row{n+1}'][0] == dic_board[f'row{n+1}'][1] == dic_board[f'row{n+1}'][2] == z:
                print('Congratulations! You have won the game!')
                return True
                
        if dic_board['row1'][0] == dic_board['row2'][1] == dic_board['row3'][2] == z:   
            print('Congratulations! You have won the game!') 
            return True
            
        if dic_board['row1'][2] == dic_board['row2'][1] == dic_board['row3'][0] == z:   
            print('Congratulations! You have won the game!')
            return True
            
    return False
                
                 
def replay():
    
    user = ''

    while user not in ['yes','no']:
        user = input('Do you want play again? Enter Yes or No ')
        if user not in ['yes','no']:
            print('Wrong input, please type yes or no! ')

    if user == 'yes':
        return True
    else:
        return False


if __name__ == '__main__':
    main()
```
